```python
from sklearn import decomposition
import pandas as pd
import numpy as np
```

```python
test = pd.read_csv(r'd:\test.csv',encoding='utf8',header=None)
test
```

```python
data = np.array(test,dtype=np.float64)
data
```

```python
#主成分pca
pca = decomposition.PCA()
pca.fit(data)
pca.transform(data)
pca.explained_variance_ratio_
```

```python
#因子分析factor
fa = decomposition.FactorAnalysis(n_components = 1)
fa.fit(data)
tran_data = fa.transform(data)
```
